```python
import glob
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def imshow_cv2plt(bgr_img, title="", axes_off=True):
    rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_img)
    plt.title(title)
    if axes_off:
        plt.axis("off")
    plt.show()
```

```python
die = np.random.randint(1,6)
die

w = 1000
h = 1000
canvas = np.zeros((w,h,3), dtype = "uint8")
```

```python
if die == 1:
    cv2.circle(canvas, (round(w/2), round(h/2)), 100, (154, 1, 254), -1)  
elif die == 2:
    cv2.circle(canvas, (850, 850), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (150, 150), 100, (154, 1, 254), -1)
elif die == 3:   
    cv2.circle(canvas, (round(w/2), round(h/2)), 100, (154, 1, 254), -1)  
    cv2.circle(canvas, (850, 850), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (150, 150), 100, (154, 1, 254), -1)
elif die == 4:  
    cv2.circle(canvas, (850, 850), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (150, 150), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (850, 150), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (150, 850), 100, (154, 1, 254), -1)
elif die == 5:
    cv2.circle(canvas, (850, 850), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (150, 150), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (850, 150), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (150, 850), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (round(w/2), round(h/2)), 100, (154, 1, 254), -1) 
else:
    cv2.circle(canvas, (850, 850), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (150, 150), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (850, 150), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (150, 850), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (850, 500), 100, (154, 1, 254), -1)
    cv2.circle(canvas, (150, 500), 100, (154, 1, 254), -1)
    
imshow_cv2plt(canvas)
```

```python
#contour time
gray = cv2.cvtColor(canvas, cv2.COLOR_BGR2GRAY)
_, threshed = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(threshed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
number = len(contours)
if number ==1:
    print("Eye'm looking at you")
elif number ==2:
    print("Tutu")
elif number ==3:
    print("I C 3 TREE")
elif number ==4:
    print("我死了")
elif number ==5:
    print('Flying 5')
else:
    print("6 let's pick up sticks")
```

```python
#contour testing
gray = cv2.cvtColor(canvas, cv2.COLOR_BGR2GRAY)
_, threshed = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)
plt.imshow(threshed, cmap="gray")
plt.axis("off")
plt.title("thresholded")
plt.show()
```

```python
contours, _ = cv2.findContours(threshed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(canvas, contours, -1, (0, 255, 0), 10)
imshow_cv2plt(canvas, "contours found")
```
